```python
import math
```

```python
def pri_not(k):
  o=0
  for i in range(1,k+1):
    if(k%i==0):
      o+=1
      if(o>2):
        return 1
        break
```

```python
def primRoots(modulo):
    roots = []
    required_set = set(num for num in range (1, modulo) if math.gcd(num, modulo) == 1)
    for g in range(1, modulo):
        actual_set = set(pow(g, powers) % modulo for powers in range (1, modulo))
        if required_set == actual_set:
            roots.append(g)           
    return roots
```

```python
P=int(input("enter the value  of P:"))
while(True):
  if pri_not(P)!=1:
    break
  else:
    P=int(input("enter the value  of P:"))
k=primRoots(P)
print("Choose one primitive root from list:",k)
alph=int(input("Enter your choice:"))
while(True):
  if alph in k:
    break
  else:
    print("Choose one element from this list:",k)
    alph=int(input("Enter your choice:"))
print("USER A")
Xa=int(input("enter your private key:"))
while(True):
  if Xa<P:
    break
  else:
    print("Key should less than P")
    Xa=int(input("enter your private key:"))
Ya=(alph**Xa)%P
print("USER B")
Xb=int(input("enter your private key:"))
while(True):
  if Xb<P:
    break
  else:
    print("Key should less than P")
    Xb=int(input("enter your private key:"))
Yb=(alph**Xb)%P
keyA=(Yb**Xa)%P
keyB=(Ya**Xb)%P
print("Secret key of A is:",keyA)
print("Secret key of B is:",keyB)
if(keyA==keyB):
  print("Key Exchange Successful")
```
