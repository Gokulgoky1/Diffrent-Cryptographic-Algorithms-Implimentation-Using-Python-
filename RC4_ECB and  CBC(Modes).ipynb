```python
#covert string to list
def Convert(string):
  list1=[]
  list1[:0]=string
  return list1
```

```python
#function to convert binary to hexadecimal
def bin_hex(bin_num):
  stri=int(bin_num,2)
  return (format(stri,'X'))
```

```python
#function to convert decimal to binary format
def deci_bin(bin):
  stri=int(bin,10)
  return format(stri,'0>8b')
```

```python
def char_to_pos(array):
  k=[]
  for i in array:
    w=char_position(i)
    k.append(w)
  return k
def pos_to_char(array):
  k=[]
  for i in array:
    w=w=pos_charter(i)
    k.append(w)
  return k
```

```python
def char_position(letter):
  return ord(letter)-97
def pos_charter(pos):
    return(chr(pos + 97))
```

# PRGA for CBC Decryption
